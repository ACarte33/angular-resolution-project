```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

amax, amin, dmin, dmax = np.loadtxt('./PlotData.csv', delimiter=',', skiprows=1, usecols=[1,2,3,4], unpack=True)
bnames = np.loadtxt('./PlotData.csv', delimiter=',', skiprows=1, dtype=str, usecols=[0], unpack=True)
fig = plt.gcf()
ax = plt.gca()

fig.set_size_inches(15, 8.5)
ax.set_xscale("log")
ax.set_yscale("log")
ax.invert_xaxis()
ax.set_ylabel("Distance (cm)", fontsize=30)
ax.set_xlabel("Angular Size")

labels = ['','','','','','','','','.28°','2.8°','28°']
ax.set_xticklabels(labels)


names = [x[2:-1] for x in bnames]

for i in range(0, len(amax)):
    if amax[i] == amin [i]: 
        plt.scatter(amin[i], dmax[i], s=70, c=np.random.rand(3,1))
    else:
        plt.plot([amin[i], amax[i]], [dmax[i], dmin[i]], linewidth=5)

plt.show()



#fig, ax = plt.subplots()
#fig.canvas.draw()

#labels = [item.get_text() for item in ax.get_xticklabels()]
#labels[1] = 'Testing'

#ax.set_xticklabels(labels)
```
