```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

amax, amin, dmin, dmax = np.loadtxt('./PlotData.csv', delimiter=',', skiprows=1, usecols=[1,2,3,4], unpack=True)
bnames = np.loadtxt('./PlotData.csv', delimiter=',', skiprows=1, dtype=str, usecols=[0], unpack=True)
fig = plt.gcf()
ax = plt.gca()

fig.set_size_inches(20, 8.5)
fig.suptitle('Angular Sizes in Astronomy', fontsize=30)
ax.set_xscale("log")
ax.set_yscale("log")
ax.invert_xaxis()
ax.set_ylabel("Distance (cm)", fontsize=20)
ax.set_xlabel("Angular Size", fontsize=20)
ax.set_xlim([108000, .0001])

locs = [.001,.01,.5,1,10,30,60,600, 1800,3600,36000,108000]
labels = ['1mas','.01\"','.5\"','1\"','10\"','30\"','1\'','10\'', '30\'','1°','10°','30°']
plt.xticks(locs, labels, fontsize=15)


#plt.plot([.05, .05], [10**12, 10**24], c='red', linewidth=2)
plt.vlines(x=0.05, linewidth=2, color='r', ymin=10**12, ymax=10**23)
plt.vlines(x=0.1, linewidth=2, color='r', ymin=10**12, ymax=10**23)
plt.vlines(x=0.14, linewidth=2, color='r', ymin=10**12, ymax=10**23)
plt.vlines(x=0.6, linewidth=2, color='r', ymin=10**12, ymax=10**23)
plt.vlines(x=1, linewidth=2, color='r', ymin=10**12, ymax=10**23)

#labeling the lines
plt.text(.057, 10**23.8, 'Hubble', fontsize=12, rotation=45)
plt.text (.12, 10**23.6,'JWST', rotation=45, fontsize=12)
plt.text (.17, 10**24, 'APO SPIcam',rotation=45, fontsize=9)
plt.text (.65, 10**23.6, 'Keck',rotation=45, fontsize=12)
plt.text (1.2, 10**23.4, 'VLA',rotation=45, fontsize=12)


names = [x[2:-1] for x in bnames]

for i in range(0, len(amax)):
    if amax[i] == amin [i]: 
        plt.scatter(amin[i], dmax[i], s=70, c=np.random.rand(3,1))
        ax.annotate(names[i], xy=(amin[i], dmax[i]), xycoords='data',
            xytext=(0, 10), textcoords='offset points', fontsize=14,
            bbox=dict(boxstyle="round", fc="0.8", alpha=.5),
                )
    else:
        plt.plot([amin[i], amax[i]], [dmax[i], dmin[i]], linewidth=5, solid_capstyle='round', alpha=.5, label=names[i])

plt.legend(loc=4, fontsize='x-large')
plt.show()


```
